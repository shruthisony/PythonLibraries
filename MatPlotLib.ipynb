**MatPlot Library of Python**

**When we need to use which type of charts with the given data..**

1.How the data is composed?
- Pie charts or Dough charts are used

2.How the values relate to each other?
- Bubble charts or Heat maps

3.How the values compare to each other?

  a. Among categories:
- Bar charts or Column charts or Grouped column charts

b.Across the time:
- Line charts or subplots of line charts

4.How the data is distributed?
- Scatter plots or Box plot or Histogram charts

**Bar Graph**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,5))
prod_cat = np.array(['Furniture','Technology','Office Supplies'])   #x
sales = np.array([100,300,200])        #y
plt.bar(prod_cat,sales,color=['cyan','hotpink','yellow'],edgecolor='black',width=0.5,align='edge')   #color=['cyan','hotpink','yellow'] for diff colors
plt.title("Bar Graph",fontdict = {
    'fontsize' : 20,
    'fontweight' : 10,
    'color' : 'green'
})
plt.xlabel('Product Categories',fontdict = {
    'fontsize' : 10,
    'fontweight' : 5,
    'color' : 'blue'
})
plt.ylabel('Sales',fontdict = {
    'fontsize' : 10,
    'fontweight' : 5,
    'color' : 'blue'
})
tick_values = np.arange(0,500,100)
tick_lables = ['0H','1H','2H','3H','4H']
plt.yticks(tick_values,tick_lables)
tick_xlabels = [i.upper() for i in prod_cat]
plt.xticks(prod_cat,tick_xlabels)
plt.show()
```

```python
plt.figure(figsize=(5,5))
prod_cat = np.array(['Furniture','Technology','Office Supplies'])
sales = np.array([100,300,200])
plt.barh(prod_cat,sales,color=['cyan','hotpink','yellow'],align='edge',edgecolor='black')   #color=['cyan','hotpink','yellow'] for diff colors
plt.title("Bar Graph",fontdict = {
    'fontsize' : 20,
    'fontweight' : 10,
    'color' : 'green'
})
plt.xlabel('Sales',fontdict = {
    'fontsize' : 10,
    'fontweight' : 5,
    'color' : 'blue'
})
plt.ylabel('Product Categories',fontdict = {
    'fontsize' : 10,
    'fontweight' : 5,
    'color' : 'blue'
})
tick_values = np.arange(0,500,100)
tick_lables = ['0H','1H','2H','3H','4H']
plt.xticks(tick_values,tick_lables)
tick_xlabels = [i.upper() for i in prod_cat]
plt.yticks(prod_cat,tick_xlabels)
plt.show()
```

**Scatter Plot**

```python
sales = np.array([1,2,6,3,7,2,8,4,9,6])
profit = np.array([100,300,500,200,150,400,350,300,200,110])
prod_cat = np.array(['A','B','C','A','A','B','C','B','A','C'])
country = np.array(['Ind','US','India','UK','US','India','UK','India','UK','US'])
plt.scatter(profit,sales,alpha=0.5)
plt.show()
```

```python
plt.scatter(profit[prod_cat == "A"], sales[prod_cat == "A"],
            c= 'Green', alpha= 0.7, s = 50, label="A" )

plt.scatter(profit[prod_cat == "B"], sales[prod_cat == "B"],
            c= 'Yellow', alpha= 0.7, s = 50, label="B" )

plt.scatter(profit[prod_cat == "C"], sales[prod_cat == "C"],
            c= 'Cyan', alpha= 0.7, s = 50, label="C" )
plt.legend()
plt.show()
```

**Line Chart**

```python
y = np.random.randint(1,100, 50)
plt.plot(y, 'cyan', marker = 'o') #can also give like this
#If you specify the colour and marker separately, then you will get a line plot with the points marked
```

```python
xpoints = np.array(['Aaaa','Bbbb','Cccc','Dddd','Eeee','Ffff','Gggg','Hhhh','Iiii','Jjjj'])
ypoints = np.array([3,8,1,10,5,7,2,8,10,4])

#plt.plot(xpoints,ypoints)   #default graph
plt.plot(xpoints,ypoints, 'red', ls = '--',color = 'black',linewidth = 2, marker = 'o',ms =7, mec = 'black', mfc = 'yellow')
plt.xticks(rotation=90)    #can apply to y-axis also as plt.yticks(rotation=y)

for xy in zip(xpoints, ypoints):
    plt.annotate(text = "{}".format(xy), xy = xy,  textcoords='data')

plt.show()
#mc -- marker color
#mec -- marker edge color
#mfc -- marker face color
#ls -- line style
```

```python
x = [1,2,3]
y = [2,5,8]

plt.subplot(1,2,1)        #(1,2,1) -- 1 represents no of rows, 2 represents no of columns, 1 represents the graph position
plt.plot(x,y)

x = [5,9,3]
y = [15,5,6]

plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
years = np.array([2012,2013,2014,2015])
sales_1 = np.array([100,200,300,400])
sales_2 = np.array([200,400,500,300])
sales_3 = np.array([150,300,500,600])
sales_4 = np.array([200,500,700,300])
sales_5 = np.array([100,300,200,400])

fig,ax = plt.subplots()
#fig, ax = plt.subplots(): It initiates a figure that will be used to compromise multiple graphs in a single chart
s1 = ax.plot(years,sales_1,label='Sales1')
s2 = ax.plot(years,sales_2,label='Sales2')
s3 = ax.plot(years,sales_3,label='Sales3',ls='--')
s4 = ax.plot(years,sales_4,label='Sales4',ls='-.')
s5 = ax.plot(years,sales_5,label='Sales5',ls='dotted')
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots(2,3)
s1 = ax[0][0].plot(years,sales_1,label='Sales1')

s2 = ax[0][1].bar(years,sales_2,label='Sales2')

s3 = ax[0][2].scatter(years,sales_3,label='Sales3')

s4 = ax[1][0].bar(years,sales_4,label='Sales4')

s5 = ax[1][1].plot(years,sales_5,label='Sales5')

s6 = ax[1][2].bar(years,sales_1,color='cyan')
s7 = ax[1][2].bar(years,sales_5,color='pink')
#ax[1][2].legend(s6,s7),('Sales_1,Sales_5')

fig.set_size_inches(10,5,forward=True)   #increases the size of charts in subplots
```

**Seaborn**

Seaborn is Python library to create statistical graphs easily. It is built on top of matplotlib and closely integrated with pandas.

_Functionalities of Seaborn_ :

- Dataset oriented API
- Analysing univariate and bivariate distributions
- Automatic estimation and plotting of  linear regression models
- Convenient views for complex datasets
- Concise control over style
- Colour palettes


```python
#import the necessary libraries
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
x=np.array([2,6,3,7,5,9,7,8,2,6,1])
sns.distplot(x,bins=15)    #distplot -- distribution plot; kde -- the density curve on the graph
plt.show()
```

```python
sns.distplot(x,kde=True,bins=15,vertical = True,color='pink')   #distplot -- distribution plot; kde -- the density curve on the graph
plt.show()            #vertical graph in pink color
```

```python
plt.figure(figsize=(5,5))
#Check all the styling options
sns.set_style('dark')                   #set_style -- to set the style of background
sns.distplot(x,bins=20,color='g')       #white, dark, whitegrid, darkgrid, ticks
plt.show()
```

```python
#Check all the styling options
sns.set_style('whitegrid')                   #set_style -- to set the style of background
sns.distplot(x,bins=20,color='g')       # white, dark, whitegrid, darkgrid, ticks
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('dark_background')          #default is also used.
sns.distplot(x,bins=20,color='cyan')
plt.show()
```

```python
plt.style.use('default')          #default is also used.
sns.distplot(x,bins=20,color='cyan')
plt.show()
```

```python
plt.style.use('ggplot')
sns.distplot(x,bins=20)
plt.show()
```

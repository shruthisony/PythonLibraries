**Series**

```python
a = np.array([10,20,30,40,50])   #Creating a Series
a1 = pd.Series(a)
print(a1)
```

```python
b = np.array([10,20,30,40])    #Creating a series with column names
b1 = pd.Series(b,index=['a','b','c','d'])
print(b1)
print(b1[0],' ',b1[3])
```

```python
c = {'a':[10,40],'b':20,'c':30,'d':40}   #Creating a series with input as a dictionary ; index names are keys automatically
c1 = pd.Series(c)
print(c1)
print(c1[0:3])
```

```python
dict1 = {"Sub":['Python','C','Java'],'rank':[1,2,3]}     #Creating a series with input as a dictionary with different data types
s1 = pd.Series(dict1)
print(s1)
print(s1[0][1])
```

```python
s = [2,'A']    #Creating a series with input as a list with different data types
a1 = pd.Series(s)
print(a1)
```

```python
s1 = pd.Series([1,2,3,4,5])   #Adding 2 series together
s2 = pd.Series([1,2,3])
s3 = s1+s2
print(s3)
```

```python
s3 = pd.Series(12)   #Creating a series with input as a tuple
print(s3)
```

**Merge() and Concat() in DataFrame**

```python
a = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})    #Same indices of A are mapped to B and C correspondingly.
pd.merge(a,b,on='A')
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})    #Different Indices
pd.merge(a,b,on='A')
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})    #Inner Join of 2 DataFrames
pd.merge(a,b,how='inner',on='A')
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.merge(a,b,how='left',on='A')        #Left Join of 2 DataFrames
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.merge(a,b,how='right',on='A')        #Right Join of 2 DataFrames
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.merge(a,b,how='outer',on='A')      #Outer Join of 2 DataFrames
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.merge(a,b,how='outer',on='A',indicator=True)     #Outer Join of 2 DataFrames with indicator
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.merge(a,b,left_index=True,right_index=True)
```

```python
a = pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.concat([a,b],axis=0)               #Clubbing both dataframes according to axis (row-wise)
```

```python
a = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
b = pd.DataFrame({'A':[1,2,3,4],'C':[9,10,11,12]})
pd.concat([a,b],axis=1)         #Clubbing both dataframes according to axis (column-wise)
```

**GroupBy()**

```python
import pandas as pd
var = pd.DataFrame({'Name':['a','a','b','b','c','c','d','d'],
                    'Sub1':[12,13,11,10,9,2,5,7],
                    'Sub2':[1,13,14,10,19,2,25,7]})
print(var)
```

```python
var_group = var.groupby('Name')    #Group by() function is used according to Name column
print(var_group)
```

```python
for x,y in var_group:
  print(x)    #Header names
  print(y)    #Data
  print()
```

```python
print(var_group.get_group('a'))    #to obtain a specific group using get_group()
print(var_group.get_group('b'))
```

```python
print(var_group.min())    #Some Statistical Functions can also be applied on grouped data
print(var_group.max())
print(var_group.mean())
print(var_group.sum())
print(var_group.count())
```

Sometimes feeling cold is more than about low temperatures; a windy day can also make you cold. A factor called the chill factor can be used to quantify the cold based on the wind speed and the temperature. The formula for the chill factor is given by


$ WCI = (10 * \sqrt{v} - v + 10.5) .(33 - T_{m}) $

v is the speed of the wind and $ T_{m} $ is the minimum temperature

Add a column for WCI and find the month with the lowest WCI.

````python
from math import sqrt
def wci(x):
    velocity = x['WindGustSpeed']
    minTemp = x['MinTemp']
    return ((10 * sqrt(velocity) - velocity + 10.5)*(33-minTemp))
````

````python
data['WCI'] = data.apply(wci,axis=1)
````

Add a new column `subs_country` where each value denotes the number of subscribers from the country of the of the corresponding user

Solution Strategy:

- First we need to perform a groupby aggregation to get the number of subscribers in each country
- After that, we need to "merge" these values with each user record from the original DataFrame

Though we can use the `pd.merge` method to combine the result of a groupby aggregation with the original DataFrame, a better approach would be to use the `transform` method.

Using this method, we get a single pandas Series which contains the groupby aggregation corresponding to each row, that can be later added to the original DataFrame. This method avoids any memory intensive merge operations and is generally faster.

Reference link --> https://stackoverflow.com/questions/30244952/how-do-i-create-a-new-column-from-the-output-of-pandas-groupby-sum

````python
users['issub'] = users.groupby(by=['country'])['is_subscriber'].transform('sum')
users
````
